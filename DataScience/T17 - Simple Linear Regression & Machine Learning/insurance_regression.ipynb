## Simple Linear Regression on an insurance dataset

The objective of this notebook is to determine how age affects insurance costs, and make prediction.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Import the data
df = pd.read_csv("insurance.csv")
df.head()
```

```python
# The iloc function allows use to select rows from the dataframe for plotting.  
x = df.iloc[:,:1].values
y = df.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
# Plot a scatter plot with age on the X-axis and charges on the Y-axis.
plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Using linear_model.LinearRegression() from sklearn, fit a model to
# your data, and make predictions on the data.

# Fit the model
charges_model = LinearRegression()
charges_model.fit(x,y)
```

```python
# Plot the data and model with the best-fit line
y_pred = charges_model.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Variable for age to predict charges
age_to_predict = [[40]]

print(f"Charges for age {age_to_predict[0][0]} should be: {charges_model.predict(age_to_predict)}")


age_to_predict = [[75]]

print(f"Charges for age {age_to_predict[0][0]} should be: {charges_model.predict(age_to_predict)}")
```

```python
# Extrapolate the linear model to a age outside of the known data.
x_pred = np.append(x, age_to_predict).reshape(-1,1)
y_pred = charges_model.predict(x_pred)

plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```
